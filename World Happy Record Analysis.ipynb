## **The ratio of happiness score in region**

```python
hr15.head()
```

It is noticing that the total dataset (hr) is merging the country and region columns in one. So I applying the dataframe from dataset of 2015 for extracting the Region column.

```python
region_list=list(hr15['Region'].unique())
region_happiness_ratio=[]
for each in region_list:
    region=hr15[hr15['Region']==each]
    region_happiness_rate=sum(region['Happiness Score'])/len(region)
    region_happiness_ratio.append(region_happiness_rate)
    
data=pd.DataFrame({'region':region_list,'region happiness ratio':region_happiness_ratio})
new_index=(data['region happiness ratio'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)

sorted_data
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=sorted_data['region'], y=sorted_data['region happiness ratio'], palette=sns.cubehelix_palette(len(sorted_data['region']),reverse=True))
plt.xticks(rotation= 90)
plt.xlabel('Region')
plt.ylabel('Region Happiness Ratio')
plt.title('Happiness rate by region')
plt.show()
```

## **The elements contribute high rank of happiness score for 10 most happy countries**

```python
f, axs = plt.subplots(3,2,figsize = (13,13))

gdp_sort = hr.sort_values(by = ['GDP'], ascending = False)
gdp_sort = gdp_sort.head(10)

social_sort = hr.sort_values(by = ['Social support'], ascending = False)
social_sort = social_sort.head(10)

health_sort = hr.sort_values(by = ['Healthy life expectancy'], ascending = False)
health_sort = health_sort.head(10)

freedom_sort = hr.sort_values(by = ['Freedom to make life choices'], ascending = False)
freedom_sort = freedom_sort.head(10)

gene_sort = hr.sort_values(by = ['Generosity'], ascending = False)
gene_sort = gene_sort.head(10)

corr_sort = hr.sort_values(by = ['Perceptions of corruption'], ascending = False)
corr_sort = corr_sort.head(10)

ax=sns.barplot(x="Country", y="GDP", data=gdp_sort, ax = axs[0,0], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Social support", data=social_sort, ax = axs[0,1], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Healthy life expectancy", data=health_sort, ax = axs[1,0], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Freedom to make life choices", data=freedom_sort, ax = axs[1,1], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Generosity", data = gene_sort, ax = axs[2,0], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Perceptions of corruption", data=corr_sort, ax = axs[2,1], palette = "bright")
ax.tick_params(axis = 'x', rotation = 45)

```
